```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# carregar dataset
df = pd.read_csv("../data/raw/train.csv")

# primeiras linhas
print(df.head())

# informações gerais
print(df.info())

# estatísticas descritivas
print(df.describe())

# valores nulos
print(df.isnull().sum())
```

```python
# Sobreviventes por sexo
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Sobrevivência por sexo")
plt.show()

# Sobreviventes por classe
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Sobrevivência por classe")
plt.show()

# Distribuição de idade por sobrevivência
sns.histplot(data=df, x="Age", hue="Survived", multiple="stack", bins=20)
plt.title("Distribuição de Idade")
plt.show()
```

Mulheres sobreviveram mais que homens (código de conduta “mulheres e crianças primeiro”).

Passageiros de 1ª classe sobreviveram mais.

Crianças tiveram taxa maior de sobrevivência.
